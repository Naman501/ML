Data Collection & Processing 

```python
#loading the csv data to a Pandas Dataframe
gold_data = pd.read_csv('/kaggle/input/gold-price-data/gld_price_data.csv')
```

Correlation:
1. Positive Correlation
2. Negative Correlation


```python
correlation = gold_data.corr()
```

```python
#constructing  a heatmap to understand the correlation 
plt.figure(figsize = (10,10))
sns.heatmap(correlation , cbar=True  ,fmt = '.3f' , annot =True , annot_kws = {'size':8},cmap ='viridis' )
```

```python
#correlation values of GLD
print(correlation['GLD'])
```

```python
# checking the distribution of the GLD 
# sns.histplot(gold_data['GLD'], color='pink')
sns.distplot(gold_data['GLD'], color='green')
```

Splitting the Features and Target

```python
X = gold_data.drop(['GLD'],axis =1)
Y = gold_data['GLD']
```

```python
print(Y)
print(X)
```

Splitting into Training Data and Test Data


```python
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size =0.2 , random_state= 2)
```

Model Training: Random Forest Regressor

```python
rf_regressor = RandomForestRegressor(n_estimators=100)

#trainig the model

rf_regressor.fit(X_train ,Y_train)
```

Model Evaluation


```python
#prediction on Test Data
test_data_prediction = rf_regressor.predict(X_test)
```

```python
print(test_data_prediction
     )
```

```python
# R square error

error_score = metrics.r2_score(Y_test,test_data_prediction)
```

```python
print("R squared Error : ",error_score)
```

Compare the Actual Values and Predicted Values in a Plot

```python
Y_test = list(Y_test)
```

```python
plt.plot(Y_test,color="Orange",label = 'Actual Values')
```

```python
plt.plot(Y_test,color="Orange",label = 'Actual Values')
plt.plot(test_data_prediction,color="pink",label='Predicted Value')
plt.title('Actual Price v/s Predicted Price')
plt.xlabel('Number of Values')
plt.ylabel('GLD Price')
plt.legend()
plt.show()
```
